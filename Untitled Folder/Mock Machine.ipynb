# KMeans 

```python
# create a copy and scale  x
airline_df= cust_df.copy()
X = airline_df.iloc[:, 1:].values

X_scaled = scale(X)
```

## Elbow Plot

```python
# define a range of values of k to evaluate for inertia and empty list
ks = range(1, 11)
inertias = []

# for each value for k
for k in ks:
  
  # fit and run the model,
  km = KMeans(n_clusters=k)
  km.fit(X_scaled)
  inertias.append(km.inertia_)
```

```python
# and plot the results
plt.plot(ks, inertias)
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

- From the graph the k = 6

# Silhouette score

```python
# define a range of values of k to evaluate for silhouette score and empty list
ks = range(2, 11)
sils = []


# for each value for k
for k in ks:

  # fit and run the model,
  km = KMeans(n_clusters=k)
  km.fit(X_scaled)
  sils.append(silhouette_score(X_scaled, km.labels_))
```

```python
# and plot the results
plt.plot(ks, sils)
plt.xlabel('no of clusters')
plt.ylabel('silhouette scoare')
plt.show()
```

```python
# Visualize different silhouette scores
# Save silhouette scores for various cluster numbers in list
silhouette_scores = []
for i in range(2, 11):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(X_scaled)
  silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))
# Visualize scores
plt.plot(range(2, 11), silhouette_scores, marker = '.')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score');
```

-From this silhouette the best value is 2

```python
# Fit kmeans model with 2 clusters and display the silhouette score
kmeans_option1 = KMeans(n_clusters = 2, random_state =42)
kmeans_option1.fit(X_scaled)
silhouette_score(X_scaled, kmeans_option1.labels_)
```
