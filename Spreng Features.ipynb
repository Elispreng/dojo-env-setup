# Load data

```python
# load pandas
import pandas as pd
```

# Prepare the data

```python
# drop the 'casual' and 'registered'
bike_df.drop(['casual', 'registered'], axis=1, inplace = True)
```

```python
#check that columns are dropped
bike_df.head()
```

```python
# check the column types
bike_df.info()
```

```python
# transform the datetime column into correct datatype
bike_df['datetime'] = pd.to_datetime(bike_df['datetime'])
bike_df.info()
```

# Transform the datetime column 

## Create three columns: month, day, and hour

```python
# create 3 columns and create column types of object
# pull out the month
bike_df['month (numeric)'] = bike_df['datetime'].dt.month
bike_df['month (name)'] = bike_df['datetime'].dt.month_name()
#df2['day of month'] = df2['date'].dt.day
#df2['day of week (numeric)'] = df2['date'].dt.weekday
# df2['day of week (name)'] = df2['date'].dt.day_name()
bike_df.head()
```

```python
# look at column types
bike_df.info()
```

```python
# pull out the day of the week
bike_df['day of month'] = bike_df['datetime'].dt.day
bike_df['day of week (numeric)'] = bike_df['datetime'].dt.weekday
bike_df['day of week (name)'] = bike_df['datetime'].dt.day_name()
bike_df.head()
```

```python
# pull out the hour from datetime column
bike_df['hour'] = bike_df['datetime'].dt.hour
bike_df.head()
```

```python
# check the column types
bike_df.info()
```

```python
# transform the hour column into an object
bike_df= bike_df.astype({'hour': object})
bike_df.info()
```

## Drop datetime and season columns

```python
#drop columns and check 
bike_df.drop(['datetime', 'season'], axis=1,inplace = True)
bike_df.head()
```

# Transform temp column to fahreheit

```python
#formula for farenheit
# (temp * 1.8) + 32
```

```python
# use lambda function for temp column
bike_df['temp'] = bike_df['temp'].apply(lambda x: (x*1.8) + 32)
bike_df.head()
```

```python
# use lambda function for atemp column
bike_df['atemp'] = bike_df['atemp'].apply(lambda x: (x*1.8) + 32)
bike_df.head()
```

```python
# round  floats for temp and atemp columns
bike_df['temp'] = bike_df['temp'].round(0).astype(int)
bike_df['atemp'] = bike_df['atemp'].round(0).astype(int)
```

```python
bike_df.head()
```

# Create new column for variance

```python
# experiment with code
bike_df['temp_variance'] = bike_df['temp']- bike_df['atemp']
```

```python
# drop atemp column
bike_df.drop['atemp']
```
